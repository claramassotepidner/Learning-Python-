```python
# 1.3.2. Obtenha os 10 anos cheios anteriores de dados
years = [2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010, 2009]

weather_last10y = weather[weather['Year'].isin(years)]

print(weather_last10y)
```

```python
# Plote um gráfico de séries "sobreposto" que permita avaliar a evolução das temperaturas máximas (Tmax) ao longo dos últimos 10 anos de dados
#registrados.
#1.4.1. Para cada ano, defina uma cor e um marcador diferente.
import matplotlib.pyplot as plt 

##
# Pegando o numeração dos meses avaliados
x_periodo = weather_last10y['Month'].unique()

# Pegando os dados de temperaturas
y_temperaturas_2009 = weather_last10y[weather_last10y['Year'] == 2009].Tmax
y_temperaturas_2010 = weather_last10y[weather_last10y['Year'] == 2010].Tmax
y_temperaturas_2011 = weather_last10y[weather_last10y['Year'] == 2011].Tmax
y_temperaturas_2012 = weather_last10y[weather_last10y['Year'] == 2012].Tmax
y_temperaturas_2013 = weather_last10y[weather_last10y['Year'] == 2013].Tmax
y_temperaturas_2014 = weather_last10y[weather_last10y['Year'] == 2014].Tmax
y_temperaturas_2015 = weather_last10y[weather_last10y['Year'] == 2015].Tmax
y_temperaturas_2016 = weather_last10y[weather_last10y['Year'] == 2016].Tmax
y_temperaturas_2017 = weather_last10y[weather_last10y['Year'] == 2017].Tmax
y_temperaturas_2018 = weather_last10y[weather_last10y['Year'] == 2018].Tmax


plt.rcdefaults()
import matplotlib.pyplot as plt

# Define uma cor para cada linha
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan', 'magenta', 'yellow', 'brown', 'gray']

plt.plot(x_periodo, y_temperaturas_2009, label="Tmax em 2009", marker='s', color=colors[0])
plt.plot(x_periodo, y_temperaturas_2010, label="Tmax em 2010", marker='p', color=colors[1])
plt.plot(x_periodo, y_temperaturas_2011, label="Tmax em 2011", marker='*', color=colors[2])
plt.plot(x_periodo, y_temperaturas_2012, label="Tmax em 2012", marker='h', color=colors[3])
plt.plot(x_periodo, y_temperaturas_2013, label="Tmax em 2013", marker='+', color=colors[4])
plt.plot(x_periodo, y_temperaturas_2014, label="Tmax em 2014", marker='x', color=colors[5])
plt.plot(x_periodo, y_temperaturas_2015, label="Tmax em 2015", marker='D', color=colors[6])
plt.plot(x_periodo, y_temperaturas_2016, label="Tmax em 2016", marker='d', color=colors[7])
plt.plot(x_periodo, y_temperaturas_2017, label="Tmax em 2017", marker='|', color=colors[8])
plt.plot(x_periodo, y_temperaturas_2018, label="Tmax em 2018", marker='_', color=colors[9])
plt.title("Controle de Temperatura Máxima dos últimos 10 anos ")
plt.ylabel("Temperatura")
plt.xlabel("Meses")
plt.grid()
plt.xticks(x_periodo)
plt.legend( bbox_to_anchor=(1.02, 1));
```

```python
# Qual o comportamento você identifica quando visualiza esse gráfico?
# 1.4.2.1. Existe uma tendência de aumento ou diminuição da temperatura máxima?
# Ao comparar os últimos 10 anos o comportamento das temperaturas máximas se manteve praticamente igual ao longo dos anos 

#1.4.2.2. Observando o gráfico é possível dizer que está ocorrendo um aquecimento da temperatura? (Sim/Não) e por quê?
# Não é possível dizer que há um aquecimento de temperatura porque o comportamento das temperaturas máximas se manteve 
# parecido entre os anos. Todos os anos apresentam um comportamento de estarem com uma temperatura mais baixa nos meses
# de Janeiro, Fevereiro, Novembro e Dezembro e mais alta nos meses de Junho, Julho e Agosto
# 1.4.2.3. Qual mês e ano foi o mais quente no período em análise?
# O mês mais quente foi o de Julho pois ele concentrou os maiores valores de temperatura máxima 
```
