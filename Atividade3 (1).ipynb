```python
# Importe, para um DataFrame Pandas, o conteúdo do arquivo CSV
import pandas as pd
jogadores = pd.read_csv("Jogadores.csv", encoding="ISO-8859-1" , delimiter = ";")
```

```python
# Mostre o conteúdo do DataFrame criado
df_jogadores = pd.DataFrame(jogadores)
```

```python
# Remova do DataFrame os jogadores que possuem o salário nulo
df_jogadores.dropna()
```

```python
# Crie uma nova coluna, salario_anual, contendo o salário anual de todos os jogadores
df_jogadores['salario_anual'] = df_jogadores['salario_do_jogador'] * 12
df_jogadores
```

```python
# Atualize o nome dos jogadores, passando todos os caracteres para maiúsculo
df_jogadores["nome_do_jogador"] = df_jogadores.nome_do_jogador.str.upper()
df_jogadores
```

```python
# Mostre os dados estatísticos dos valores numéricos do DataFrame (comando
# describe). Descreva as informações sobre os salários dos jogadores, obtidas através
# desses dados

df_jogadores.salario_do_jogador.describe

# Sobre o salário dos quatro jogadores, o valor mínimo é de 100.000,00,
# a média salarial é 312.500,00 e o valor máximo é 850.000,00. Um dos valores é nulo
```

```python
#Faça consultas no DataFrame para exibir:
#a. Nome e time dos jogadores que possuem salário acima de R$ 200000,00
df_jogadores[df_jogadores['salario_do_jogador'] > 200000.0] [['nome_do_jogador', 'nome_time_jogador']]
```

```python
# b. Nome e salário dos jogadores dos times de Minas Gerais (estado MG)
df_jogadores[df_jogadores['nome_estado_jogador'] == "MG"] [['nome_do_jogador', 'salario_do_jogador']]
```

```python
# c. Nome e time dos jogadores cujo nome contenha a letra 'u'

df_jogadores[df_jogadores.nome_do_jogador.str.contains('U')] [['nome_do_jogador', 'nome_time_jogador']]

```

```python
#d. Nome, salário e time dos jogadores, ordenados pelo salário em ordem decrescente

df_jogadores.sort_values(by = ['salario_do_jogador'], ascending = False )[['nome_do_jogador','salario_do_jogador' ,'nome_time_jogador']]
```

```python
# e. Nome, salário e time dos jogadores, ordenados pelo nome do time em ordem
# crescente, e em seguida pelo salário em ordem decrescente
df_jogadores.sort_values(by = ['nome_time_jogador', 'salario_do_jogador'], ascending = [True , False] )[['nome_do_jogador','salario_do_jogador' ,'nome_time_jogador']]
```

```python
# f. A quantidade de jogadores por time

df_jogadores['nome_time_jogador'].value_counts()
```

```python
# g. A média salarial por time

df_jogadores.groupby('nome_time_jogador')['salario_do_jogador'].mean()
```
